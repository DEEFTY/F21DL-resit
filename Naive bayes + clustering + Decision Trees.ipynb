### 1.3 Filtering滤波处理

在图像识别机器学习任务中，滤波处理是一个常用的预处理步骤，以下是一些滤波处理的一些优点：  

- **去除噪声**：图像可能受到各种类型的噪声污染，例如高斯噪声、椒盐噪声等。这些噪声会干扰对图像信息的正确理解和分析。通过应用适当的滤波器，可以降低或去除这些噪声，提高图像质量和后续处理的准确性。  
- **平滑图像**：有时候，图像中的细节过多或过于突出可能对后续处理任务造成困扰。使用平滑滤波器可以模糊图像，减少细节，使得图像更加平滑和均匀，以便更好地处理和分析。  
- **特征增强**：某些情况下，我们希望突出或增强图像中的某些特定特征，例如边缘、纹理等。通过应用合适的滤波器，可以增强这些特征，使得它们更加明显和易于识别。  

需要注意的是，滤波处理应根据具体的任务和图像特点进行选择和调整。不同的滤波器和参数会产生不同的效果，需要根据实际情况进行优化和调试。  

滤波处理可以改善**图像质量、增强特征、去除噪声**等，从而为后续的图像识别机器学习任务提供更好的输入数据。  
In image recognition machine learning tasks, filtering is a common preprocessing step. Here are some advantages of filtering:  

- **Noise removal**: The image may be polluted by various types of noise, such as Gaussian noise, salt and pepper noise, etc. These noises can interfere with proper understanding and analysis of image information. By applying appropriate filters, these noises can be reduced or removed, improving image quality and subsequent processing accuracy.  
- **Smooth Image**: Sometimes too much or too much detail in an image can be a problem for subsequent processing tasks. Using a smoothing filter can blur the image, reduce details, and make the image smoother and more uniform for better processing and analysis.  
- **Feature Enhancement**: In some cases, we want to highlight or enhance some specific features in the image, such as edges, textures, etc. By applying suitable filters, these features can be enhanced to make them more obvious and recognizable.  

It should be noted that the filtering process should be selected and adjusted according to specific tasks and image characteristics. Different filters and parameters will produce different effects, which need to be optimized and debugged according to the actual situation.  

Filtering can improve** image quality, enhance features, remove noise**, etc., thereby providing better input data for subsequent image recognition machine learning tasks.

#### **中值滤波去噪**  

中值滤波：中值滤波是一种非常简单但有效的去噪方法。它通过将每个像素的值替换为它周围像素值的中值来去除噪声。  

可以使用OpenCV库中的cv2.medianBlur函数来应用中值滤波。  
#### **Median filter denoising**  

Median filtering: Median filtering is a very simple but effective method of denoising. It removes noise by replacing the value of each pixel with the median value of the pixel values around it.  

Median filtering can be applied using the cv2.medianBlur function in the OpenCV library.

```python
import warnings 
warnings.filterwarnings('ignore') 
print(X.shape,"\n",X[0,:,:,0])
X[0,:,:,0].reshape(-1,X.shape[1]*X.shape[1])

X_changed=pd.DataFrame(X[0,:,:,0].reshape(-1,X.shape[1]*X.shape[1]))

for i in range(1,X.shape[0]):
    tmp = pd.DataFrame(X[i,:,:,0].reshape(-1,X.shape[1]*X.shape[1]))
    #将 df2 的行追加到现有 DataFrame 的末尾append rows of df2 to end of existing DataFrame
#     X_changed.loc[i]=tmp
    X_changed = X_changed.append(tmp)
    X_changed = X_changed.reset_index().drop(["index"],axis=1)
#     X_changed=X_changed.reset_index()


X_changed["label"]=y
df_changed = X_changed.copy()
df_changed
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, roc_curve, auc
import numpy as np
import matplotlib.pyplot as plt
from itertools import combinations
# 引入必要的库
import numpy as np
import matplotlib.pyplot as plt
from itertools import cycle
from sklearn import svm, datasets
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
from scipy import interp
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import numpy as np
import matplotlib.pyplot as plt



X = df_changed.drop(["label"],axis=1)
y = df_changed.label


# 洗牌和分割训练集和测试集shuffle and split training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2,random_state=0)

# 创建朴素贝叶斯分类器Create a Naive Bayes Classifier
clf = GaussianNB()

# 拟合模型fit model
clf.fit(X_train, y_train)

# 在测试集上进行预测Make predictions on the test set
y_pred = clf.predict(X_test)

# 计算混淆矩阵Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# 计算分类报告Calculate Classification Report
report = classification_report(y_test, y_pred)
print("Classification Report:")
print(report)


```

```python
# 在测试集上进行预测Make predictions on the test set
y_pred = clf.predict_proba(X_test)
y_pred

classes=[0,1,2]

y_test_all = label_binarize(y_test, classes=[0,1,2])

# 在测试集上进行预测Make predictions on the test set
y_pred = clf.predict_proba(X_test)
y_pred

y_score_all=y_pred
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(len(classes)):
    fpr[i], tpr[i], thresholds = roc_curve(y_test_all[:, i],y_score_all[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

roc_auc

# micro-average ROC curve（方法一）
fpr["micro"], tpr["micro"], thresholds = roc_curve(y_test_all.ravel(),y_pred.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
roc_auc



# macro-average ROC curve 方法二）

all_fpr = np.unique(np.concatenate([fpr[i] for i in range(len(classes))]))

mean_tpr = np.zeros_like(all_fpr)
for i in range(len(classes)):
    mean_tpr += interp(all_fpr, fpr[i], tpr[i])
# 求平均计算ROC包围的面积AUC(Calculate the average area AUC surrounded by ROC
mean_tpr /= len(classes)
fpr["macro"] = all_fpr
tpr["macro"] = mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

#画图部分drawing part
plt.figure()
plt.plot(fpr["micro"], tpr["micro"],'k-',color='y',
         label='ROC curve micro-average(AUC = {0:0.4f})'
               ''.format(roc_auc["micro"]),
          linestyle='-.', linewidth=3)

plt.plot(fpr["macro"], tpr["macro"],'k-',color='k',
         label='ROC curve macro-average(AUC = {0:0.4f})'
               ''.format(roc_auc["macro"]),
          linestyle='-.', linewidth=3)
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.1,1.1])
plt.ylim([-0.1,1.1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc="lower right")
plt.grid(linestyle='-.')  
plt.grid(True)
plt.show()
```

### 2.4 Elbow Method确定最优聚类个数  (Elbow Method to determine the optimal number of clusters)  

确定最优聚类个数的一种常用方法是使用K均值聚类（K-means clustering）的“手肘法”（Elbow Method）。  

手肘法基于聚类结果的变异程度来选择最优的聚类个数。当聚类个数增加时，每个样本点与其所属聚类中心的距离会减小，导致总体的误差减小。然而，随着聚类个数继续增加，每个新增的聚类中心所带来的误差减少逐渐减缓，最终达到一个拐点。这个拐点称为“手肘”，它反映了在聚类个数增加过程中，聚类效果的显著改善程度逐渐降低。  
A common method to determine the optimal number of clusters is to use the "Elbow Method" of K-means clustering.  

The elbow method selects the optimal number of clusters based on the degree of variation of the clustering results. When the number of clusters increases, the distance between each sample point and its cluster center will decrease, resulting in a decrease in the overall error. However, as the number of clusters continues to increase, the error reduction brought by each new cluster center gradually slows down, and finally reaches an inflection point. This inflection point is called "elbow", which reflects that the degree of significant improvement in clustering effect gradually decreases as the number of clusters increases.  


```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler

X = df_changed.iloc[:,:81]
# 标准化数据normalized data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 设置不同的聚类个数进行试验Experiment with different numbers of clusters
k_values = range(1, 9)
sse = []

for k in k_values:
    kmeans = KMeans(n_clusters=k,random_state=123)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)

# 绘制手肘法图形Drawing the Elbow Method Graphics
plt.plot(k_values, sse, 'bx-')
plt.xlabel('Number of Clusters')
plt.ylabel('SSE (Sum of Squared Errors)')
plt.title('Elbow Method with Standardized Data')
plt.show()
```

不难看出，拐点在$k=4$附近，这说明该数据的最优聚类个数为4个  
The optimal number of clusters for this data is 4
